```python
import sys
with open('romeoandjuliet.txt','r') as file:
    textfile=file.read().split()
# Create a Dictionary 
dct=dict()
for word in textfile:
    word=word.lower()
    word=word.strip('!')
    word=word.strip(',')
    word=word.strip('.')
    word=word.strip(':')
    word=word.strip(';')
    word=word.strip('\'')
    word=word.strip('\"')
    word=word.strip('?')
    word=word.strip("")
    word=word.strip('[')
    word=word.strip(" ")
    word=word.strip("(")
    word=word.strip(")")
    word=word.strip("-")

    if word in dct:
        dct[word]+=1
    else:
        dct.update({word : 1})


                   
```

```python
# store the value of top 10 words that most frequent appearance
import collections
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
word_counter=collections.Counter(dct)

list_commom_word=['and','the','i','to','a','of','my','that','is','in','at']
for word,count in word_counter.most_common(10):
        print(word,":",count)
most= word_counter.most_common(10)
df_most = pd.DataFrame(most, columns = ['Word', 'Count'])
df_most.plot.bar(x='Word',y='Count')


# plot the 10 words that least be written in the book
for word,count in word_counter.most_common()[:-10-1:-1] :
        print(word,":",count)
least = word_counter.most_common()[:-10-1:-1]
df_least = pd.DataFrame(least, columns = ['Word', 'Count'])
df_least.plot.bar(x='Word',y='Count')
plt.show()
```

The most ten frequent words are preposition and article. They are the key parameters for constructing a full sentence. However, for the noun and verbs can interchange when we use it, their appearance frequency would be less. 

```python
word_count=sorted(word_counter.values())
word_count.reverse()
rank=range(len(dct))
plt.loglog(word_count,rank)
plt.xlabel('log[word count]')
plt.ylabel('log[word rank]')
plt.title('log[word count] vs log[word rank] in Shakespeare Romeo and Juliet')
plt.show()
```

As the word counting gets to higher, the word rank is higer as well
